# User Interface (Python)

```python
# this code was expanded from the code provided in Session 12.2 "Expert System with Shell"
'''
Set up prolog by importing and setting up Prolog interpreter to use it with Python.
'''

prolog = Prolog() # Global handle to interpreter

retractall = Functor("retractall")
known = Functor("known",3)

def write_py(X):
    '''
    It outputs the value to the standard output.
    It ensures the output is shown by using sys.stdout.flush

    Parameters:
    - X (str): The value to be printed.

    Returns:
    - True (bool): Indicates that the function finished successfully.
    '''
    print(str(X))
    sys.stdout.flush()
    return True

def read_py(A,V,Y):
    '''
    Reads a boolean input from the user about a restaurant attribute and then unifies
    with a Prolog variable.

    Parameters:
    - A (str): The attribute being asked about.
    - V (Variable): Not use in this function, but part of Prolog.
    - Y (Variable): The variable that unifies the user's response.

    Returns:
    - True (bool): If the input was properly read and unified.
    - False (bool): If an error occurred an the input was not properly read and unified.
    '''
    if isinstance(Y, Variable):
        formatted_A = str(A).replace("_", " ")
        print(f"Should the restaurant have {formatted_A}? enter yes/no.")
        response = input()
        while response != "yes" and response != "no":
            print("Invalid input. Please enter yes or no.")
            response = input()
        print("Your input:", response, "\n")
        Y.unify(response)
        return True
    else:
        return False

def read_menu_py(A, V, option_list):
    '''
    It allows the user to select a preference from a list of options for one of the
    restaurant attributes. This shows a numbered list of options and asks the user
    to choose one by entering the corresponding number. It also validates the input
    and unifies the option selected by the user with a Prolog variable.

    Parameters:
    - A (str): The attribute for which preferences are being set.
    - V (Variable): The variable that will unify with the response.
    - option_list (list): A list of the options we have for the attribute.

    Returns:
    - True (bool): If the input was properly read and unified.
    - False (bool): If an error occurred an the input was not properly read and unified.
    '''
    if isinstance(V, Variable):
        formatted_A = str(A).replace("_", " ")
        print(f"Please chose your preference for {formatted_A}. \nEnter the number: ")
        for num, option in enumerate(option_list, start=1):
            print(f"{num}. {option}")
        user_choice = int(input())
        # validate input
        while user_choice not in range(1, len(option_list)+1):
            print("Invalid input. Please enter a number in the range.")
            user_choice = int(input())
        chosen_list_member = str(list(option_list)[user_choice-1])
        print("Your input:", chosen_list_member, "\n")
        V.unify(chosen_list_member)
        return True
    else:
        return False


# Set the arity for the foreign functions to allow for their registration
write_py.arity = 1
read_py.arity = 3
read_menu_py.arity = 3

# Register foreign function from Python in Prolog
registerForeign(read_py)
registerForeign(write_py)
registerForeign(read_menu_py)

# Create a temporary file with the KB in it
(FD, KB_file_name) = tempfile.mkstemp(suffix='.pl', text = "True")
with os.fdopen(FD, "w") as text_file:
    text_file.write(test_KB)
prolog.consult(KB_file_name) # open the KB for consulting
os.unlink(KB_file_name) # Remove the temporary file

# Remove all the known facts in the imported prolog KB
call(retractall(known))
print("Welcome to the restaurant recommender.\n")
restaurant = [s for s in prolog.query("restaurant(X).")]
if restaurant:
  print("We recommend:")
  print(restaurant[0]['X'])
else:
  print("We're sorry, we couldn't find a restaurant matching your preferences.")
  print("You can try running the recommender again with different preferences.")
```
